# Evaluation of our Model

1) Using a Webcam

2) Using a video feed

Algorithm Used: VGG16 Keras Library

## 1) Using a Webcam

```python
from keras.models import load_model
from time import sleep
from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
import cv2
import numpy as np

face_classifier = cv2.CascadeClassifier(r'inputs/haarcascade_frontalface_default.xml')
classifier = load_model(r'inputs/VGG16-keras.h5')

emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)
        roi_gray = gray[y:y + h, x:x + w]

        # Resize and repeat channels to match ResNet input shape
        roi_resized = cv2.resize(roi_gray, (224, 224))
        roi_rgb = cv2.cvtColor(roi_resized, cv2.COLOR_GRAY2RGB)
        roi_rgb = np.expand_dims(roi_rgb, axis=0)

        if np.sum([roi_gray]) != 0:
            roi = roi_rgb.astype('float') / 255.0

            prediction = classifier.predict(roi)[0]
            label = emotion_labels[prediction.argmax()]
            label_position = (x, y)
            cv2.putText(frame, label, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        else:
            cv2.putText(frame, 'No Faces', (30, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Emotion Detector', frame)
    
    # Check if the 'x' key is pressed
    key = cv2.waitKey(1)
    if key == ord('x') or key == ord('X'):
        break

cap.release()
cv2.destroyAllWindows()
```

## 2) Using a video feed

```python
from keras.models import load_model
from keras.preprocessing import image
import cv2
import numpy as np

# Load the pre-trained VGG16 model
classifier = load_model('inputs/VGG16-keras.h5')

# Define emotion labels
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# Load the Haar cascade classifier for face detection
face_classifier = cv2.CascadeClassifier('inputs/haarcascade_frontalface_default.xml')

# Open the video feed
cap = cv2.VideoCapture('inputs/video-feed.mp4')

while True:
    # Read a frame from the video feed
    ret, frame = cap.read()
    
    # Break the loop if the end of the video is reached
    if not ret:
        break
    
    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the frame
    faces = face_classifier.detectMultiScale(gray)
    
    # Process each detected face
    for (x, y, w, h) in faces:
        # Extract the region of interest (ROI) containing the face
        roi_gray = gray[y:y+h, x:x+w]
        
        # Resize the ROI to match the input shape of the VGG16 model
        input_image = cv2.resize(roi_gray, (224, 224))
        
        # Convert the resized image to RGB format
        input_image_rgb = cv2.cvtColor(input_image, cv2.COLOR_GRAY2RGB)
        
        # Convert the image to an array and expand the dimensions
        input_image_array = image.img_to_array(input_image_rgb)
        input_image_array = np.expand_dims(input_image_array, axis=0)
        
        # Normalize the input image
        input_image_array /= 255.0
        
        # Predict the emotion using the VGG16 model
        prediction = classifier.predict(input_image_array)[0]
        
        # Get the predicted emotion label
        label = emotion_labels[np.argmax(prediction)]
        
        # Draw a rectangle around the face and display the predicted emotion label
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    # Display the processed frame
    cv2.imshow('Emotion Detector', frame)
    
    # Check if the 'x' key is pressed
    key = cv2.waitKey(1)
    if key == ord('x') or key == ord('X'):
        break

# Release the video feed and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
